```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Lets first see if its even installed right, and what the version is

try:
    nx.__version__
    print("NetworkX is installed, version:", nx.__version__)
except AttributeError:
    print("NetworkX is installed, but version information is unavailable.")
```

```python
#Lets try a basic graph
G = nx.Graph()
G.add_node(1)
G.add_node(2)


G.add_edge(1, 2)


print("Nodes of the graph:", G.nodes())
print("Edges of the graph:", G.edges())
```

```python
#Lets make a basic krackhardt kite
G = nx.krackhardt_kite_graph()

#Lets graph it
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, edge_color='black', node_size=500, font_size=12) #Lets draw the kite

plt.title("Krackhardt Kite Graph")
plt.show()
```
