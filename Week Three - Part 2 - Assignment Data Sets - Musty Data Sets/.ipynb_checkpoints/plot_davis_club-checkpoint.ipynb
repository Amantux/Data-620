#Start of supplied info
# Davis Club

Davis Southern Club Women

Shows how to make unipartite projections of the graph and compute the
properties of those graphs.

These data were collected by Davis et al. in the 1930s.
They represent observed attendance at 14 social events by 18 Southern women.
The graph is bipartite (clubs, women).


```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import bipartite
import pandas as pd

#Start of supplied python code to make the graphs for womens
G = nx.davis_southern_women_graph()
women = G.graph["top"]
clubs = G.graph["bottom"]

print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))

# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
print()
print("#Friends, Member")
for w in women:
    print(f"{W.degree(w)} {w}")

# project bipartite graph onto women nodes keeping number of co-occurrence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print()
print("#Friend meetings, Member")
for w in women:
    print(f"{W.degree(w, weight='weight')} {w}")

pos = nx.spring_layout(G, seed=648)  # Seed layout for reproducible node positions
nx.draw(G, pos)
plt.show()


#End of supplied python code
```

```python
W_unweighted = bipartite.projected_graph(G, women)

W_weighted = bipartite.weighted_projected_graph(G, women)

degree_centrality = nx.degree_centrality(W_weighted)
betweenness_centrality = nx.betweenness_centrality(W_weighted)
closeness_centrality = nx.closeness_centrality(W_weighted)
eigenvector_centrality = nx.eigenvector_centrality(W_weighted)

#Create a DataFrame to display centrality measures
centrality_df = pd.DataFrame({
    'Degree Centrality': degree_centrality,
    'Betweenness Centrality': betweenness_centrality,
    'Closeness Centrality': closeness_centrality,
    'Eigenvector Centrality': eigenvector_centrality
})
centrality_df


```

```python
#Visualize the original bipartite graph
pos = nx.spring_layout(G, seed=648)  #Seed layout for reproducible node positions
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Davis Southern Club Women Bipartite Graph")
plt.show()

```

```python
#Visualize the weighted projected graph with centrality measures
plt.figure(figsize=(15, 10))
pos_projected = nx.spring_layout(W_weighted, seed=648)

#Draw the network nodes
node_color = list(degree_centrality.values())
nx.draw_networkx_nodes(W_weighted, pos_projected, node_color=node_color, cmap=plt.cm.viridis, node_size=500, alpha=0.8)

#Draw Edges
nx.draw_networkx_edges(W_weighted, pos_projected, alpha=0.3)

#Label
nx.draw_networkx_labels(W_weighted, pos_projected, font_size=12)

plt.title("Davis Southern Club Women Network")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis), label='Degree Centrality')
plt.show()
```

```python
#filter to just nodes
women_nodes = {n for n, d in G.nodes(data=True) if d['bipartite'] == 0}
event_nodes = set(G) - women_nodes

#degree
degree_centrality = nx.degree_centrality(G)
event_degree_centrality = {n: degree_centrality[n] for n in event_nodes}

#betweeness
betweenness_centrality = nx.betweenness_centrality(G)
event_betweenness_centrality = {n: betweenness_centrality[n] for n in event_nodes}

#closeness
closeness_centrality = nx.closeness_centrality(G)
event_closeness_centrality = {n: closeness_centrality[n] for n in event_nodes}

eigenvector_centrality = nx.eigenvector_centrality_numpy(G)
event_eigenvector_centrality = {n: eigenvector_centrality[n] for n in clubs}


event_attendance = {event: len(list(G.neighbors(event))) for event in event_nodes}
event_metrics = pd.DataFrame({
    'Degree Centrality': event_degree_centrality,
    'Betweenness Centrality': event_betweenness_centrality,
    'Closeness Centrality': event_closeness_centrality,
    'Eigenvector Centrality': event_eigenvector_centrality,
    'Number of Unique Attendees': event_attendance,
})
```

```python
# Create bar plots for each metric
metrics = ['Degree Centrality', 'Betweenness Centrality', 'Closeness Centrality','Eigenvector Centrality', 'Number of Unique Attendees']

for metric in metrics:
    plt.figure(figsize=(12, 6))
    plt.bar(event_metrics.index, event_metrics[metric], color='skyblue')
    plt.xlabel('Events')
    plt.ylabel(metric)
    plt.title(f'{metric} of Events')
    plt.xticks(rotation=90)
    plt.show()
```

What can you infer about the relationships between (1) the women, and (2) the social events?  

I can assume that there is a core group of women that have a high degree of connection with each other, ie a core group of friends or socialites. From there, theres an outer fringe that has a connection to a socialite but isn't a primary node.
